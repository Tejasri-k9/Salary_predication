# Salary Predication
Problem Statement: Predicting salaries based on years of experience using a dataset containing this information.

1.Data Preprocessing: Load dataset, handle missing values.

2.Feature Extraction: Extract years of experience (X) and corresponding salaries (y).

3.Model Selection: Choose Linear Regression for its simplicity and effectiveness in predicting continuous target variables.

4.Model Training: Split data, train Linear Regression model on training set.

5.Prediction: Use trained model to predict salaries on testing set.

6.Visualization: Plot actual salaries against years of experience, overlay with regression line.

7.Model Evaluation: Assess model performance using metrics like Mean Squared Error and R-squared.
    
Using Linear Regression is chosen for its simplicity, interpretability, and ability to capture the linear relationship between years of experience and salaries.
